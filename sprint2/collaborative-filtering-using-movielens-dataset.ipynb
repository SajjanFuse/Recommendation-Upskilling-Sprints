## Making a collaborative filtering model that takes (user_id, movie_id) and gives (1-5) 

```python
from sklearn.metrics import mean_squared_error

def rmse(y_true, y_pred):
    return np.sqrt(mean_squared_error(y_true, y_pred))

# defining a baseline model that always returns 3
def baseline(user_id, movie_id):
    return 3.0 
```

```python
# collaborative filtering model attempts to predict the ratings of the movies 
def score(cf_model):
    """
    Returns the error of the model with test dataset
    """
    id_pairs = zip(X_test['user_id'], X_test['movie_id'])
    y_pred = np.array([cf_model(user, movie) for (user,movie) in id_pairs])
    y_true = np.array(X_test['rating'])
    return rmse(y_true, y_pred) 
```

```python
# making a user based cf 
# matrix between movies and users 
ratings_matrix = X_train.pivot_table(values = 'rating', index='user_id', columns = 'movie_id')
print(ratings_matrix.shape) 
print(type(ratings_matrix))
print(ratings_matrix.head(2))
```

```python
def cf_user_mean(user_id, movie_id):
    if movie_id in ratings_matrix:
    # mean of all the ratings given to the movie
        mean_rating = ratings_matrix[movie_id].mean()
    else:
    # default to a rating of 3.0
        mean_rating = 3.0
    return mean_rating
# print(user_id, movie_id)
score(cf_user_mean)
```

```python
# using consine similarity 
# changing NaNN values to 0, since cosine similarity does not work with NaN 
r_matrix_dummy = ratings_matrix.copy().fillna(0)
# Import cosine_score
from sklearn.metrics.pairwise import cosine_similarity
#Compute the cosine similarity matrix using the dummy ratings matrix
cosine_sim = cosine_similarity(r_matrix_dummy, r_matrix_dummy)
#Convert into pandas dataframe
cosine_sim = pd.DataFrame(cosine_sim, index=ratings_matrix.index,
columns=ratings_matrix.index)
cosine_sim.head(10)
```

```python
def cf_user_wmean(user_id, movie_id):
    #Check if movie_id exists in r_matrix
    if movie_id in r_matrix_dummy:
        #Get the similarity scores for the user in question with every other user
        sim_scores = cosine_sim[user_id]
        #Get the user ratings for the movie in question
        m_ratings = r_matrix_dummy[movie_id]
        #Extract the indices containing NaN in the m_ratings series
        idx = m_ratings[m_ratings.isnull()].index
        #Drop the NaN values from the m_ratings Series
        m_ratings = m_ratings.dropna()
        #Drop the corresponding cosine scores from the sim_scores series
        sim_scores = sim_scores.drop(idx)
        #Compute the final weighted mean
        wmean_rating = np.dot(sim_scores, m_ratings)/ sim_scores.sum()
    else:
        #Default to a rating of 3.0 in the absence of any information
        wmean_rating = 3.0
    return wmean_rating
score(cf_user_wmean)
```

```python
# merging users_df and X_train
merged_df = pd.merge(X_train, users_df)
merged_df.head(2)
```

```python
gender_mean = merged_df[['movie_id', 'sex', 'rating']].groupby(['movie_id', 'sex'])['rating'].mean()
print(len(merged_df['movie_id'].unique()))
gender_mean.head(), gender_mean.info 
```

```python
# users_df = users_df.set_index('user_id')
#Gender Based Collaborative Filter using Mean Ratings
def cf_gender(user_id, movie_id):
    #Check if movie_id exists in r_matrix (or training set)
    if movie_id in ratings_matrix:
        #Identify the gender of the user
        gender = users_df.loc[user_id]['sex']
        #Check if the gender has rated the movie
        if gender in gender_mean[movie_id]:
            #Compute the mean rating given by that gender to the movie
            gender_rating = gender_mean[movie_id][gender]
        else:
            gender_rating = 3.0
    else:
        #Default to a rating of 3.0 in the absence of any information
        gender_rating = 3.0
    return gender_rating
score(cf_gender)
```

```python
# using both gender and occupation
gen_occ_mean = merged_df[['sex', 'rating', 'movie_id', 'occupation']].pivot_table( values='rating', index='movie_id', columns=['occupation', 'sex'], aggfunc = 'mean')
gen_occ_mean.head(2)
```

```python
def cf_gen_occ(user_id, movie_id):
    #Check if movie_id exists in gen_occ_mean
    if movie_id in gen_occ_mean.index:
        #Identify the user
        user = users_df.loc[user_id]
        #Identify the gender and occupation
        gender = user['sex']
        occ = user['occupation']
        #Check if the occupation has rated the movie
        if occ in gen_occ_mean.loc[movie_id]:
            #Check if the gender has rated the movie
            if gender in gen_occ_mean.loc[movie_id][occ]:
                #Extract the required rating
                rating = gen_occ_mean.loc[movie_id][occ][gender]
                #Default to 3.0 if the rating is null
                if np.isnan(rating):
                    rating = 3.0
                return rating
    return 3.0

score(cf_gen_occ)
```

```python
def recommend_movies(user_id, gen_occ_mean):
    user = users_df.loc[user_id]
    gender = user['sex']
    occupation = user['occupation']
#     print(user_id, gender, occupation)
    
    movies_ratings = gen_occ_mean[occupation, gender]
    top_ten = movies_ratings.dropna().sort_values(ascending=False).head(20)
    
    return top_ten
```
